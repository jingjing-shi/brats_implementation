```python
dest_folder = '/Users/Jingjing/project 2/brats2015/TRAINING/'
```

```python
# convert all images in different modalities to png
from shutil import copyfile
training_path = '/Users/Jingjing/project 2/brats2015/train'
for label in os.listdir(training_path):
    if label.startswith('.'):
        pass
    else:
        label_path = os.path.join(training_path,label)
        for pat in os.listdir(label_path):
            if pat.startswith('.'):
                pass
            else:
                pat_path = os.path.join(label_path,pat)
                for file in os.listdir(pat_path):
                    if 'OT.' in file:
                        modality = 'Seg'
                    elif 'MR_T2.' in file:
                        modality = 'T2'
                    elif 'MR_T1.' in file:
                        modality = 'T1'
                    elif 'MR_T1c.' in file:
                        modality = 'T1c'
                    elif 'MR_Flair.' in file:
                        modality = 'Flair'
                    output_folder = os.path.join(dest_folder,label,modality)
                    file_path = os.path.join(pat_path,file)
                    if modality == 'Seg':
                        copyfile(os.path.join(pat_path,file),os.path.join(output_folder,file))
                    else:
                        image_array, image_header = load(file_path)
                        nx,ny,nz = image_array.shape
                        total_slices = image_array.shape[2]
                        for current_slice in range(0,total_slices):
                            data = image_array[::,current_slice]
                            image_name = output_folder+'/'+'.'.join(file.split('.')[:-1])+ "_z" + "{:0>3}".format(str(current_slice+1))+'.png'
                            imageio.imwrite(image_name, data)
                           
```
